Colab T4 training notebook. Clone repo, install deps, verify CUDA, train PPO, save artifacts.

```python
import os
import sys
import subprocess
from pathlib import Path

REPO_URL = "https://github.com/TheHarshal30/Reinforcement-Learning-Course-Project.git"
REPO_DIR = Path("Reinforcement-Learning-Course-Project")

if not REPO_DIR.exists():
    subprocess.run(["git", "clone", REPO_URL], check=True)
else:
    subprocess.run(["git", "-C", str(REPO_DIR), "pull"], check=True)

%cd Reinforcement-Learning-Course-Project
subprocess.run([sys.executable, "-m", "pip", "install", "-q", "-r", "requirements.txt"], check=True)
subprocess.run([
    sys.executable,
    "-m",
    "pip",
    "install",
    "-q",
    "torch",
    "--index-url",
    "https://download.pytorch.org/whl/cu121",
], check=True)
```

```python
import torch
print("torch:", torch.__version__)
print("cuda available:", torch.cuda.is_available())
if torch.cuda.is_available():
    print("gpu:", torch.cuda.get_device_name(0))
    print("cuda version:", torch.version.cuda)
```

```python
from api_rate_limiter_rl.config import EnvConfig, PPOConfig
from api_rate_limiter_rl.experiments import run_experiments

# Colab-friendly preset. Increase train_episodes for a longer run.
env_cfg = EnvConfig(episode_length=180, seed=7)
ppo_cfg = PPOConfig(
    hidden_size=32,
    actor_lr=0.005,
    critic_lr=0.008,
    device="auto",
    gamma=0.97,
    lam=0.93,
    clip_eps=0.2,
    update_epochs=4,
    minibatch_size=32,
    train_episodes=120,
    eval_episodes=12,
    seed=7,
)

print("device preset:", ppo_cfg.device)
result = run_experiments("artifacts", env_cfg=env_cfg, ppo_cfg=ppo_cfg)
print(result["report_path"])
```
